```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
datafile = 'Netflix subscription fee Dec-2021.csv'
```

```python
df = pd.read_csv(datafile)
df
```

```python
plt.figure(figsize=(30,10))

sns.histplot(x = df['Country_code'],
 y = df['Cost Per Month - Premium ($)'])
plt.show()
```

```python
plt.figure(figsize=(30,10))

sns.histplot(x = df['Country_code'],
 y = df['Cost Per Month - Standard ($)'])
plt.show()
```

```python
plt.figure(figsize=(30,10))

sns.barplot(x = df['Country_code'],
 y = df['Cost Per Month - Standard ($)'])
plt.show()
```

```python
plt.figure(figsize=(30,10))

sns.barplot(x = df['Country_code'],
 y = df['Total Library Size'])
plt.show()
```

```python
sns.histplot(df['Cost Per Month - Basic ($)'], bins=25)
plt.show
```

```python
sns.boxplot(df['Cost Per Month - Standard ($)'])
plt.show
```

```python
sns.boxplot(df['Cost Per Month - Premium ($)'])
plt.show
```

```python
b = pd.pivot_table(df,
               index = 'Country',
               values = ['Cost Per Month - Basic ($)','Cost Per Month - Standard ($)', 'Cost Per Month - Premium ($)' ],
              )
b.head
```

```python
b.plot(kind = 'barh', figsize=(30,10) )
```

```python
Bar Chart que te dice en que pais se cobra mas ya sea basic, standard o premium
```

```python
c = pd.pivot_table(df,
               index = 'Country',
               values = ['Total Library Size', 'No. of TV Shows', 'No. of Movies' ],
              )
c.head
```

```python
c.plot(kind = 'barh', figsize=(30,10) )
```
